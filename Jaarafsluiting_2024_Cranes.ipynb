```python
# Optional: convert pages to a temporary PDF
def extract_pages_to_temp_pdf(input_pdf_path, output_pdf_path, start_page, end_page):
    doc = fitz.open(input_pdf_path)
    new_doc = fitz.open()
    for i in range(start_page - 1, end_page):  # 0-based indexing
        new_doc.insert_pdf(doc, from_page=i, to_page=i)
    new_doc.save(output_pdf_path)
    new_doc.close()

# Paths
input_pdf_path = x
temp_pdf_path = y


# Step 1: Extract pages 4 to 6 into a new PDF
extract_pages_to_temp_pdf(input_pdf_path, temp_pdf_path, 4, 6)

# Step 2: Use tabula to extract tables
# Note: This requires Java and tabula-py to be installed.
tables = tabula.read_pdf(temp_pdf_path, pages='all', multiple_tables=True)

# Step 3: Combine all tables into a single DataFrame
combined_df = pd.concat(tables, ignore_index=True)

# Optional: Clean up temporary file
os.remove(temp_pdf_path)

# Display the combined dataframe
print(combined_df)
```

```python
Balans = combined_df
Balans.rename(columns={'Unnamed: 0':'Rubriek'}, inplace=True)
Balans.rename(columns={'Vorig boekjaar':'Vorig_boekjaar'}, inplace=True)
Balans = Balans.drop('Unnamed: 1', axis=1)
#Balans = Balans.drop('Toel.', axis=1)
Balans = Balans.fillna(0)
```

```python
# prompt: Dataframe data gebruiken: convert columns Boekjaar_2024,  Boekjaar_2023, Boekjaar_2022 and Boekjaar_2021 in numeric columns

# List of columns to convert to numeric
columns_to_convert = ['Boekjaar_2024', 'Boekjaar_2023', 'Boekjaar_2022', 'Boekjaar_2021']
# Loop through the columns and convert them to numeric
for column in columns_to_convert:
    # Replace the dot '.' with an empty string to handle thousands separators
    data[column] = data[column].astype(str).str.replace('.', '', regex=False)
    # Convert the cleaned column to numeric, coercing errors to NaN
    data[column] = pd.to_numeric(data[column], errors='coerce')
    data
```

```python
df = data
df = df.drop('Codes', axis=1)
df = df.set_index('Rubriek')
df= df.transpose()
#data.rename(columns=data.iloc[0])
df.columns
```

```python
# Ratios
df['Liquiditeit'] = df['VLOTTENDE ACTIVA'] / df['Schulden op ten hoogste één jaar']
df['Acid_test'] = df['Vorderingen op ten hoogste één jaar'] / df['Schulden op ten hoogste één jaar'] + df['Overlopende rekeningen A'] / df['Schulden op ten hoogste één jaar']

#Solvabiliteit VV P (totaal vermogen)
df['Solvabiliteit_S'] = df['EIGEN VERMOGEN'] / df['SCHULDEN']
df['Solvabiliteit_A'] = df['EIGEN VERMOGEN'] / df['TOTAAL VAN DE ACTIVA']
df['Solvabiliteit_LT'] = df['EIGEN VERMOGEN'] / df['Schulden op meer dan één jaar']




# NA, NBA, NBK
df['Netto Actief'] = df['TOTAAL VAN DE ACTIVA'] - df['SCHULDEN'] - df["Voorzieningen voor risico's en kosten"] # Boekhoudkundig eigen vermogen
df['NBA'] = df['Vorderingen op ten hoogste één jaar'] + df['Overlopende rekeningen A'] - df["Handelsschulden"] - df["Overlopende rekeningen"]# NBA Netto Bedrijfsactiva, Vermogensstructuur op lange termijn + Winstgevendheid & waardering

df['Bedrijfs Activa'] = df['Vorderingen op ten hoogste één jaar'] + df['Overlopende rekeningen A']
df['Bedrijfs Passiva'] = df["Handelsschulden"] + df["Overlopende rekeningen"]

df['NBK'] = df['EIGEN VERMOGEN'] +  df['Schulden op meer dan één jaar'] - df["VASTE ACTIVA"]
df['NBK_1'] = df['VLOTTENDE ACTIVA']  - df["Schulden op ten hoogste één jaar"]

df['Nettokas'] = df['NBK']  - df["NBA"]


# EBIT
df['EBIT'] = df['Bedrijfswinst (Bedrijfsverlies)']  + df["Financiële opbrengsten"] - df["Financiële kosten"]
df['EBITDA'] = df['EBIT'] + df['Afschrijvingen']




# Financiële Schuld
df['Netto Financiele schuld'] = df['Financiële schulden']  - df["Geldbeleggingen"] - df["Liquide middelen"]
df['Netto Financiele schuldsgraad'] = df['Netto Financiele schuld'] / df['EIGEN VERMOGEN']
df['Netto Financiele schuld ratio'] = df['Netto Financiele schuld'] / df['EBITDA']

# Intrestdekking
df['Intrestdekking'] = df['Winst (Verlies) van het boekjaar vóór belasting'] / df['Financiële kosten'] - 1

# Schuldaflossingscapaciteit
df['Schuldaflossingscapaciteit'] = df['Netto Financiele schuld'] / df['EBITDA']

# Rentabiliteit op basis van de EV of P
df['Rentabiliteit_EV'] = df['Bedrijfswinst (Bedrijfsverlies)'] / df['EIGEN VERMOGEN']
df['Rentabiliteit_P'] = df['Bedrijfswinst (Bedrijfsverlies)'] / df['TOTAAL VAN DE PASSIVA']

#Financiele hefboom (Veel schuld → meer risico en potentieel hoger rendement)
df['Financiele_hefboom'] = df['SCHULDEN'] / df['EIGEN VERMOGEN'] # VV/EV = (R/VV)/(R/EV)
df['Financiele_hefboom_1'] = df['TOTAAL VAN DE ACTIVA'] / df['EIGEN VERMOGEN'] # VV/EV = (R/VV)/(R/EV)
df

df
```
